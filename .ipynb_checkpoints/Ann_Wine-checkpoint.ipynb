```python
import sklearn 
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("winequality-red-4.csv", delimiter=";")
data = data[1:]
X,y = data[:,0:-1], data[:,-1]
print(X)
#X = sklearn.preprocessing.normalize(X)
#print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
clf = MLPClassifier(random_state=10, max_iter=800).fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
activation = ['identity', 'logistic', 'tanh', 'relu']
alpha = [0.00001, 0.0001,0.001,0.01,0.1,1,10]
learning_rate = [0.001,0.01,0.1,1,10]
scores = np.zeros((4,7,5))
```

```python
for ind1, act in enumerate(activation):
    for ind2, alp in enumerate(alpha):
        for ind3, lr in enumerate(learning_rate):
            #print(act,alp,lr)
            clf = MLPClassifier(random_state=10, max_iter=800, activation=act, alpha=alp, learning_rate_init=lr).fit(X_train, y_train)
            scores[ind1][ind2][ind3] = clf.score(X_test, y_test)

print(np.argmax(scores))
```

```python
max_index = np.argmax(scores)

max_index_3d = np.unravel_index(max_index, scores.shape)

print(max_index_3d,scores)
print(activation[max_index_3d[0]],alpha[max_index_3d[1]],learning_rate[max_index_3d[2]])
```

```python
names = ["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]

for ind, name in enumerate(names):
    plt.scatter(X[:,ind],y)
    plt.xlabel(name)
    plt.ylabel("quality")
    plt.title(f"Scatter plot of {name} and quality") 
    plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

colors = ['Red', 'Orange', 'Green', 'Blue', "cyan", "purple"]

figure, axis = plt.subplots(2, 3)
for color, i, target_name,q,w in zip(colors, ([3,4,5,6,7,8]), ([3,4,5,6,7,8]), [0,0,0,1,1,1],[0,1,2,0,1,2]):
    axis[q,w].scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, lw=2, label=target_name, )
    axis[q,w].legend(loc="best", shadow=False, scatterpoints=1)
    axis[q,w].set_title(f"quality {i}")
plt.show()

for color, i, target_name in zip(colors, ([3,4,5,6,7,8]), ([3,4,5,6,7,8])):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, lw=2, label=target_name, )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA of Wine dataset")
plt.show()
```
